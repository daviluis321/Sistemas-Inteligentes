```python
#Questao1
import graphlab
```

```python
dados = graphlab.SFrame('train.csv')
```

```python
dados.head()
```

```python
#Questao2
mat = dados['NU_NOTA_MT'] #peso = 4
ch = dados['NU_NOTA_CH'] #peso = 2
cn = dados['NU_NOTA_CN'] #peso = 3
red = dados['NU_NOTA_REDACAO'] #peso = 5
lc = dados['NU_NOTA_LC'] #peso = 4
```

```python
mediap = ((mat*4) + (cn*3) + (ch*2) + (red*5) + (lc*4)) / (4 + 2 + 3 + 4 + 5)
```

```python
dados['media'] = mediap
```

```python
dados
```

```python
#Questao3
melhores_piores_alunos = graphlab.SFrame(dados['NU_INSCRICAO', 'media'])
```

```python
#melhores alunos
melhores_piores_alunos.topk('media', k=15, reverse=False ) 
```

```python
#piores alunos
melhores_piores_alunos.topk('media', k=15, reverse=True) 
```

```python
#Questao4
dados = dados.filter_by(None,'media',exclude=True)
dados["boa_ruim"] = dados["media"] >= 200
```

```python
dados['NU_INSCRICAO', 'boa_ruim']
```

```python
dados["boa_ruim"].show(view="Categorical")
```

```python
#Questao5
treino, teste = dados.random_split(.8, seed=0)
modelo1 = graphlab.logistic_classifier.create(treino, target='boa_ruim',
                                             features=['media'],
                                             validation_set=teste)
```

```python
#Questao6
modelo2 = graphlab.svm_classifier.create(treino, target='boa_ruim',
                                        features=['media'],
                                        validation_set=teste)
```

```python
#Questao7
resultados = teste["boa_ruim"]
predicao1 = modelo1.predict(teste)
matrix1 = graphlab.evaluation.confusion_matrix(resultados, predicao1)
matrix1
```

```python
predicao2 = modelo2.predict(teste)
matrix2 = graphlab.evaluation.confusion_matrix(resultados, predicao2)
matrix2
```
