# Importação do GraphLab

```python
import graphlab
```

# Importação da Base de Dados

```python
vendas = graphlab.SFrame('home_data.gl/')
```

```python
vendas 
```

# Seleção Estatística

```python
selecao = vendas[vendas['zipcode'] == '98178']
```

```python
selecao
```

# Preço médio de vendas

```python
print selecao['price'].mean()
```

```python
graphlab.canvas.set_target('ipynb')
selecao[('sqft_living') , 'price'].show(view='Scatter Plot')
```

# Filtro de dados

```python
filtro = vendas[(vendas['sqft_living'] > 2000) & (vendas['sqft_living'] < 4000)]
```

```python
filtro
```

# Divisão treino/teste

```python
vendas_treino, vendas_teste = vendas.random_split(.8, seed=0)
```

# Construíndo o modelo 

```python
vendas_model = graphlab.linear_regression.create(vendas_treino, target='price', features=['sqft_living'], validation_set=None)
```

# Explorando outras características das casas

```python
caracteristicas = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
```

# Explorando outras características avançadas das casas

```python
caracteristicas_avancadas = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode', 'condition', 'waterfront', 'grade', 'view', 'sqft_above', 'sqft_basement']
```

# Construíndo o modelo de características

```python
vendas_caracteristicas = graphlab.linear_regression.create(vendas_treino, target='price', 
                                  features=caracteristicas, validation_set=None)
```

# Construíndo o modelo de características avançadas

```python
vendas_caracteristicas_avancadas = graphlab.linear_regression.create(vendas_treino, target='price', features=caracteristicas_avancadas, validation_set=None)
```

# Avaliação para o modelo, características e características avançadas

```python
print vendas_model.evaluate(vendas_teste)
print vendas_caracteristicas.evaluate(vendas_teste)
print vendas_caracteristicas_avancadas.evaluate(vendas_teste)
```
